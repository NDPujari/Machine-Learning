```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel('flight_price.xlsx')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.Date_of_Journey
```

```python
df['day']=df.Date_of_Journey.str.split('/').str[0]
df['month']=df.Date_of_Journey.str.split('/').str[1]
df['year']=df.Date_of_Journey.str.split('/').str[2]
```

```python
df.info()
```

```python
df.day=df.day.astype(int)
df.month=df.month.astype(int)
df.year=df.year.astype(int)
```

```python
df.info()
```

```python
df.drop(columns='Date_of_Journey', inplace=True)
```

```python
df.columns
```

```python
df.Arrival_Time
```

```python
df['Arrival_Time'].apply(lambda x:x.split(' '))
```

```python
# want only time
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Time']
```

```python
df['Arrival_Hour']=df.Arrival_Time.str.split(':').str[0].astype(int)
df['Arrival_Min']=df.Arrival_Time.str.split(':').str[1].astype(int)
```

```python
df.drop(columns='Arrival_Time', inplace=True)
```

```python
df['Dep_Hour']=df.Dep_Time.str.split(':').str[0].astype(int)
df['Dep_Min']=df.Dep_Time.str.split(':').str[1].astype(int)
```

```python
df.drop(columns='Dep_Time', inplace=True)
```

```python
df.dtypes
```

```python
df.Total_Stops.unique()
```

```python
df.Total_Stops=df.Total_Stops.replace({'non-stop':0, '2 stops':2, '1 stop':1, '4 stops':4, '3 stops':3}).astype(int)
```

```python
df.Total_Stops.unique()
```

```python
df.dtypes
```

```python
df.drop(columns='Route', inplace=True)
```

```python
df.info()
```

```python
df.Duration
```

```python
df['Duration_hours']=df.Duration.str.split(' ').str[0].str.split('h').str[0]
```

```python
df['Duration_min']= df.Duration.str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_min']=df.Duration_min.fillna(0)
```

```python
df.Duration_min=df.Duration_min.astype(int)
```

```python
df.drop(columns='Duration', inplace=True)
```
