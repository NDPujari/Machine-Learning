Question 2: What are the types of machine learning? Describe each with one
real-world example.
-->
    Types of ML:
    1) Supervised Learning
    2) UnSupervised Learning
    3) Reinforcement Learning
    4) Semi-supervised Learning

    1) Supervised Learning: The model is trained on a labeled dataset, meaning each input has a known, correct output. The algorithm learns to map the inputs to the outputs by comparing its predictions to the correct answers and improving over time.
    Example: A spam filter is trained on emails that are already labeled as either "spam" or "not spam".

    2) UnSupervised Learning: The model is given unlabeled data and must find hidden patterns, structures, or groupings on its own. There is no "correct" answer to guide the learning process.
    Example: A retail company uses unsupervised learning to cluster its customers into different groups based on their purchasing behavior, without being given any predefined categories.

    3) Reinforcement Learning: An agent learns to make decisions by interacting with an environment through trial and error. It receives "rewards" for good actions and "penalties" for bad ones, and its goal is to develop a strategy that maximizes its total reward over time.
    Example: A self-driving car learns to navigate by receiving positive rewards for successful actions (like staying in its lane) and negative penalties for errors (like a near-collision).

    4) Semi-Supervised Learning: This approach combines both supervised and unsupervised learning. The model is trained on a small amount of labeled data and a large amount of unlabeled data, which helps improve accuracy and efficiency compared to using only one type of data.
    Example: A large photo library may have a few photos labeled with the names of the people in them. The model uses this small labeled set to help identify and label people in the rest of the unlabeled photos, using the similarities it finds.

Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.
-->
    Overfitting: Overfitting occurs when a machine learning model learns the training data too well, including the noise and specific details that are not representative of the underlying patterns. This leads to excellent performance on the training set but poor generalization to new, unseen data. An overfit model is overly complex and has high variance.

    Underfitting: Underfitting occurs when a machine learning model is too simple to capture the underlying patterns and relationships within the training data. This results in poor performance on both the training data and unseen test data. An underfit model has high bias and fails to learn the essential features of the data.

    Bias-Variance: The bias-variance tradeoff is a fundamental concept in machine learning that describes the inherent conflict between a model's bias and variance.
    Bias- Bias refers to the error introduced by approximating a real-world problem, which may be complex, with a simplified model. High bias indicates that the model makes strong assumptions and is likely to underfit.
    Variance- Variance refers to the amount that the model's predictions would change if different training data were used. High variance indicates that the model is too sensitive to the training data and is likely to overfit.

Question 4: What are outliers in a dataset, and list three common techniques for handling them.
-->
    Outliers: An outlier is a data point in a dataset that is significantly different from the other observations. Outliers can be unusually high or low values.

    Techniques for handling Outliers:
    1) Removing Outliers: This involves deleting the outlier data points from the dataset. It is a straightforward method, but should be used with caution, particularly with small datasets, as it can lead to information loss and bias. It is most appropriate when there is a strong reason to believe the outlier is an error, such as a data entry mistake.

    2) Data transformation: Mathematical transformations can reduce the impact of outliers by changing the scale of the data. This is particularly useful for skewed data, as it can make the distribution more symmetric. Methods of transformation are Log Transformation, Square Root Transformation.

    3) Capping: This method replaces extreme values with the next most extreme value that is not considered an outlier.This technique reduces the impact of outliers without removing them entirely, helping to preserve the integrity of the data distribution.

Question 5: Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.
-->
    The process of handling missing values involves identifying missing data and using imputation to replace it with estimated values, preserving the dataset's size. For numerical data, a common imputation technique is to use the mean or median of the existing values. For categorical data, the mode (the most frequent value) is a standard imputation method.

```python
''' Question 6: Write a Python program that:
● Creates a synthetic imbalanced dataset with make_classification() from
sklearn.datasets.
● Prints the class distribution. '''

from sklearn.datasets import make_classification
from collections import Counter

X, y = make_classification(
    n_samples=1000,
    n_features=10,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    weights=[0.9, 0.1],
    random_state=42
)


class_count = Counter(y)

print("Class distribution:")
for label, count in class_count.items():
    percent = (count / len(y)) * 100
    print(f"Class {label}: {count} samples ({percent:.2f}%)")
```

```python
''' Question 8: Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation.  '''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = np.random.randn(1000)

df = pd.DataFrame(data, columns=['Values'])

missing_indices = np.random.choice(df.index, 50, replace=False)
df.loc[missing_indices, 'Values'] = np.nan

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(df['Values'], bins=30, color='skyblue', edgecolor='black')
plt.title("Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

mean_value = df['Values'].mean()
df['Values'].fillna(mean_value, inplace=True)

plt.subplot(1, 2, 2)
plt.hist(df['Values'], bins=30, color='lightgreen', edgecolor='black')
plt.title("After Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
