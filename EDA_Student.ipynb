```python
# outlier detection
# to detect outlier >> distribution, boxplot, violinplot
student_grades_df=df.copy()
student_grades_df
```

```python
student_grades_df.drop(['G1', 'G2', 'G3'], axis=1)
```

```python
import numpy as np
import pandas as pd

def detect_outliers(columns):
    outlier_indices = []
    for column in columns:
        Q1 = student_grades_df[column].quantile(0.25)
        Q3 = student_grades_df[column].quantile(0.75)
        IQR = Q3 - Q1

        # mask is the boolean condition
        mask = (student_grades_df[column] >= (Q1 - 1.5 * IQR)) & (student_grades_df[column] <= (Q3 + 1.5 * IQR))
        mask = mask.to_numpy() # bool to numpy

        # find the indices of values that do not satisfy the mask condition >> that will be outlier
        false_indices = np.argwhere(~mask)
        outlier_indices.extend(false_indices) # Use extend instead of append

    return np.unique(np.concatenate(outlier_indices).ravel())
```

```python
num_cols=['age', 'absences']
detect_outliers(num_cols)
```
